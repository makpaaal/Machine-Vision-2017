```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('.\\6_26.jpg')
imGBR = pltimg.imread('.\\6_26.jpg')
imRBG = pltimg.imread('.\\6_26.jpg')
imGRB = pltimg.imread('.\\6_26.jpg')
imBRG = pltimg.imread('.\\6_26.jpg')
imBGR = pltimg.imread('.\\6_26.jpg')
fig = plt.figure(figsize=(15,10))

imGBR[:,:, 1] = im[:,:, 0]
imGBR[:,:, 2] = im[:,:, 1]
imGBR[:,:, 0] = im[:,:, 2]

imRBG[:,:, 0] = im[:,:, 0]
imRBG[:,:, 2] = im[:,:, 1]
imRBG[:,:, 1] = im[:,:, 2]

imGRB[:,:, 1] = im[:,:, 0]
imGRB[:,:, 0] = im[:,:, 1]
imGRB[:,:, 2] = im[:,:, 2]

imBRG[:,:, 2] = im[:,:, 0]
imBRG[:,:, 0] = im[:,:, 1]
imBRG[:,:, 1] = im[:,:, 2]

imBGR[:,:, 2] = im[:,:, 0]
imBGR[:,:, 1] = im[:,:, 1]
imBGR[:,:, 0] = im[:,:, 2]


plt.subplot(321)
plt.imshow(im)
plt.subplot(322)
plt.imshow(imGBR)
plt.subplot(323)
plt.imshow(imRBG)
plt.subplot(324)
plt.imshow(imGRB)
plt.subplot(325)
plt.imshow(imBRG)
plt.subplot(326)
plt.imshow(imBGR)
plt.show()
```
