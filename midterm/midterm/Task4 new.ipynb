```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
import scipy.misc as simg

from skimage import feature

im = simg.imread("./simpleimage.jpg", mode="L")

im = ndi.gaussian_filter(im, 4)

# Compute the Canny filter for three values of sigma
edges1 = feature.canny(im)
edges2 = feature.canny(im, sigma=3)
edges3 = feature.canny(im, sigma=5)

# display results
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4, figsize=(8, 3),
                                    sharex=True, sharey=True)

ax1.imshow(im, cmap=plt.cm.Reds)
ax1.axis('off')
ax1.set_title('noisy image', fontsize=10)

ax2.imshow(edges1, cmap=plt.cm.Reds)
ax2.axis('off')
ax2.set_title('Canny filter, $\sigma=1$', fontsize=10)

ax3.imshow(edges2, cmap=plt.cm.Reds)
ax3.axis('off')
ax3.set_title('Canny filter, $\sigma=3$', fontsize=10)

ax4.imshow(edges3, cmap=plt.cm.Reds)
ax4.axis('off')
ax4.set_title('Canny filter, $\sigma=5$', fontsize=10)

fig.tight_layout()

plt.show()
```
